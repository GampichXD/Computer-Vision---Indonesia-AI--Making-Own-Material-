```python
#import library
import numpy as np #for numerical operations
import matplotlib.pyplot as plt #for data visualization
from sklearn.datasets import make_moons #for generating synthetic dataset
from sklearn.model_selection import train_test_split #for splitting dataset into training and testing sets
from sklearn.tree import DecisionTreeClassifier #for creating decision tree classifier
from sklearn.metrics import accuracy_score #for calculating accuracy score

#generate synthetic dataset (dua bulan sabit)
X, y = make_moons(n_samples=300, noise=0.3, random_state=42)

#split dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

#Visualization
plt.scatter(X[:, 0], X[: , 1], c=y, cmap='bwr', edgecolors='k')
plt.title('Dataset: Dua Bulan Sabit (Klasifikasi Biner)')
plt.xlabel('Fitur 1')
plt.ylabel('Fitur 2')
plt.show()
```

```python
#Model underfitting
underfit_model = DecisionTreeClassifier(max_depth=1, random_state=42) #model dengan kedalaman maksimum 1
underfit_model.fit(X_train, y_train)

train_pred = underfit_model.predict(X_train)
test_pred = underfit_model.predict(X_test)

print("Underfitting Model (max_depth=1)")
print(f"Training Accuracy: {accuracy_score(y_train, train_pred):.3f}")
print(f"Testing Accuracy: {accuracy_score(y_test, test_pred):.3f}")
```

```python
#Model good fit
good_model = DecisionTreeClassifier(max_depth=5, random_state=42) #model dengan kedalaman maksimum 5
good_model.fit(X_train, y_train)

train_pred = good_model.predict(X_train)
test_pred = good_model.predict(X_test)

print("\nGood Fit Model (max_depth=5)")
print(f"Training Accuracy: {accuracy_score(y_train, train_pred):.3f}")
print(f"Testing Accuracy: {accuracy_score(y_test, test_pred):.3f}")
```

```python
#Model overfitting
overfit_model = DecisionTreeClassifier(max_depth=20, random_state=42) #model dengan kedalaman maksimum 20
overfit_model.fit(X_train, y_train)

train_pred = overfit_model.predict(X_train)
test_pred = overfit_model.predict(X_test)

print("\nOverfitting Model (max_depth=20)")
print(f"Training Accuracy: {accuracy_score(y_train, train_pred):.3f}")
print(f"Testing Accuracy: {accuracy_score(y_test, test_pred):.3f}")
```

```python
#Fungsi plot boundary
def plot_boundary(model, X, y, title):
    h = 0.02
    x_min, x_max = X[:, 0].min() - 0.5, X[:, 0].max() + 0.5
    y_min, y_max = X[:, 1].min() - 0.5, X[:, 1].max() + 0.5
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                         np.arange(y_min, y_max, h))
    Z = model.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    plt.contourf(xx, yy, Z, alpha=0.3, cmap='bwr')
    plt.scatter(X[:, 0], X[:, 1], c=y, cmap='bwr', edgecolors='k')
    plt.title(title)
    plt.show()

plt.figure(figsize=(15,4))
plt.subplot(131)
plot_boundary(underfit_model, X_test, y_test, "Underfitting Model (max_depth=1)")
plt.subplot(132)
plot_boundary(good_model, X_test, y_test, "Good Fit Model (max_depth=5)")
plt.subplot(133)
plot_boundary(overfit_model, X_test, y_test, "Overfitting Model (max_depth=20)")
plt.tight_layout()
plt.show()
```

```python
#Make cricles

#generate synthetic dataset (dua bulan sabit)
from sklearn.datasets import make_circles


X, y = make_circles(n_samples=300, noise=0.1, factor=0.5, random_state=42)

#split dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

#Visualization
plt.scatter(X[:, 0], X[: , 1], c=y, cmap='bwr', edgecolors='k')
plt.title('Dataset: Dua Lingkaran (Klasifikasi Biner)')
plt.xlabel('Fitur 1')
plt.ylabel('Fitur 2')
plt.show()
```

```python
#Model underfitting circles
underfit_model = DecisionTreeClassifier(max_depth=1, random_state=42) #model dengan kedalaman maksimum 1
underfit_model.fit(X_train, y_train)

train_pred = underfit_model.predict(X_train)
test_pred = underfit_model.predict(X_test)

print("Underfitting Model (max_depth=1)")
print(f"Training Accuracy: {accuracy_score(y_train, train_pred):.3f}")
print(f"Testing Accuracy: {accuracy_score(y_test, test_pred):.3f}")
```

```python
#Model good fit
good_model = DecisionTreeClassifier(max_depth=6, random_state=42) #model dengan kedalaman maksimum 5
good_model.fit(X_train, y_train)

train_pred = good_model.predict(X_train)
test_pred = good_model.predict(X_test)

print("\nGood Fit Model (max_depth=5)")
print(f"Training Accuracy: {accuracy_score(y_train, train_pred):.3f}")
print(f"Testing Accuracy: {accuracy_score(y_test, test_pred):.3f}")
```

```python
#Model overfitting
overfit_model = DecisionTreeClassifier(max_depth=20, random_state=42) #model dengan kedalaman maksimum 20
overfit_model.fit(X_train, y_train)

train_pred = overfit_model.predict(X_train)
test_pred = overfit_model.predict(X_test)

print("\nOverfitting Model (max_depth=20)")
print(f"Training Accuracy: {accuracy_score(y_train, train_pred):.3f}")
print(f"Testing Accuracy: {accuracy_score(y_test, test_pred):.3f}")
```

```python
#regularization min_samples_leaf

#Model good fit
good_model = DecisionTreeClassifier(min_samples_leaf=5, random_state=42) #model dengan kedalaman maksimum 5
good_model.fit(X_train, y_train)

train_pred = good_model.predict(X_train)
test_pred = good_model.predict(X_test)

print("\nGood Fit Model (min_samples_leaf=5)")
print(f"Training Accuracy: {accuracy_score(y_train, train_pred):.3f}")
print(f"Testing Accuracy: {accuracy_score(y_test, test_pred):.3f}")
```
